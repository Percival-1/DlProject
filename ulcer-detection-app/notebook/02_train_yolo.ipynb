```python
from ultralytics import YOLO
from pathlib import Path
import yaml
```

```python
DATA_YAML = Path('../dataset/data.yaml')
print(DATA_YAML.resolve())
with open(DATA_YAML, 'r') as f:
    print(f.read())
```

### Choose model and quick smoke training

```python
model = YOLO('yolov8n-seg.pt')
# Windows-Safe Command
model.train(
    data=str(DATA_YAML),
    epochs=100,      # Set a high ceiling
    patience=10,     # STOP if no improvement for 15 epochs
    imgsz=640,
    batch=16,
    workers=0,       # Keep this at 0 for Windows!
    device=0,
    name='ulcer_final_run'
)
```

```python
from shutil import copyfile
from glob import glob


ckpt = glob('runs/segment/ulcer_final_run/weights/best*.pt')
if ckpt:
    print('Found checkpoint:', ckpt[0])
    Path('../models/trained').mkdir(parents=True, exist_ok=True)
    copyfile(ckpt[0], '../models/trained/best.pt')
else:
    print('No checkpoint found in runs/segment/train/weights')
```
